```python
#Spliting the dataset into dependent(X) and independent(Y) dataset
Y=df.iloc[:,1].values
X=df.iloc[:,2:32].values
```

### Train - Test Split

```python
#Creating training and testing dataset by spliting 
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
#feature scaling - Standard Scalarization
from sklearn.preprocessing import StandardScaler
X_train = StandardScaler().fit_transform(X_train)
X_test = StandardScaler().fit_transform(X_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
#Gererating Models
def models(X_train,Y_train):

    #Logistic Regression
    log=LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    
    #Decision Tree Classifier 
    tree=DecisionTreeClassifier(random_state=0,criterion='entropy')
    tree.fit(X_train,Y_train)
    
    #Random Forest
    forest=RandomForestClassifier(random_state=0,criterion='entropy',n_estimators=10)
    forest.fit(X_train,Y_train)
    
    #K Neighbors Classifier 
    neighbor=KNeighborsClassifier(n_neighbors=3)
    neighbor.fit(X_train,Y_train)
    
    #Gaussian Naive Bayes
    gnb=GaussianNB()
    gnb.fit(X_train,Y_train)
    
    print('Decision Tree Accuracy : ',tree.score(X_train,Y_train))
    print('Logistic Regression Accuracy :',log.score(X_train,Y_train))
    print('Random Forest Accuracy :',forest.score(X_train,Y_train))
    print('KNeighborsClassifier Accuracy :',log.score(X_train,Y_train))
    print('Gaussian Naive Bayes :',log.score(X_train,Y_train))
    
    return tree,log,forest,neighbor,gnb
```

```python
model=models(X_train,Y_train)
```

```python
#Model Testing
from sklearn.metrics import accuracy_score,classification_report

for i in range(len(model)):
    print("Model -",i+1)
    print(classification_report(Y_test,model[i].predict(X_test)))
    print("Accuracy :",accuracy_score(Y_test,model[i].predict(X_test)))
    print("\n")
```

### Ensembling The Model

```python
from sklearn.ensemble import VotingClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import RepeatedStratifiedKFold
```

```python
estimators = list()

logistic = Pipeline([('m', LogisticRegression())])
estimators.append(('logistic', logistic))

randomforest = Pipeline([('m', RandomForestClassifier())])
estimators.append(('randomforest', randomforest))

kneighbor = Pipeline([('m',KNeighborsClassifier(n_neighbors=3))])
estimators.append(('neighbour', kneighbor))

# define the voting ensemble
ensemble = VotingClassifier(estimators=estimators, voting='hard')
```

```python
ensemble
```

```python
cv = RepeatedStratifiedKFold(n_splits=30, n_repeats=3, random_state=1)
n_scores = cross_val_score(ensemble, X, Y, scoring='accuracy', cv=cv, n_jobs=-1)
```

```python
n_scores
```

```python
n_scores.mean()
```

```python
ensemble.fit(X_train,Y_train)
```

```python
print("Model -Ensembled-VotingClassifier")
print(classification_report(Y_test,ensemble.predict(X_test)))
print("Accuracy :",accuracy_score(Y_test,ensemble.predict(X_test)))
print("\n")
```

```python
#Prediction through random forest model
p=ensemble.predict(X_test)
print("PREDICTED VALUES ".center(75,"-")+"\n\n",p)
print("\n"+"ACTUAL VALUES ".center(75,"-")+"\n\n",Y_test)
```

```python
import pickle
#Dumping the ensemble
pickle.dump(ensemble,open("Breast Cancer Predictor.pkl",'wb'))
```
